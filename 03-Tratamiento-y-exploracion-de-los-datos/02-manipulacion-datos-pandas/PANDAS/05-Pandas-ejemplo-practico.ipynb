## 1. Inspreccion inicial

```python
# para iniciar el codigo

import pandas as pd
```

```python
# crear un datafrane apartit de un diccionario
data = {
    'Nombre': ['Ana','luis','carlos','lucia','Martha','Jorge','ivan'],
    'edad': [23,35,45,25,30,28,32],
    'ciudad' : ['CDMX','guadarajala', 'montery','CDMX','Puebla','CDMX','guadarajala'],
    'salario' : [12000, 15000, 18000, 12500,14000, 13000, 15000],
    'Departamento': ['Ventas','Marketing','Ventas','Ti','Ti','Marketing','Ventas']
}
```

```python
# day a frame

df = pd.DataFrame(data)
df 
```

# 4.- filtrado de filas 

```python
# filtrar todas las filas que la ciudad seca CDMX
df[df['ciudad'] == 'CDMX']
```

```python
df[df['ciudad'] == 'CDMX'][['Nombre', 'ciudad', 'salario']]
```

```python
df_cdmx = df[df['ciudad'] == 'CDMX']
```

```python
df_cdmx[['Nombre','ciudad','salario']]
```

```python
(df['Departamento'] == 'Ti')
```

```python
# mostrar todos los nombres de la ciudad de mexico y departamento ti
df[(df['ciudad'] == 'CDMX') & (df['Departamento'] == 'Ti')]
```

### mostrar todas las filas de la ciudad de mexico o del departamento de ti

```python
df[(df['ciudad'] == 'CDMX') | (df['Departamento'] == 'Ti')]
```

```python
df.sort_values(by='salario')
```

```python
# ordena por salario de forma acendente (A-Z) (0-9)
df.sort_values(by='salario', ascending=True)
```

```python
# ordena por salario de forma sendente (Z-A) (9-0)
df.sort_values(by='salario', ascending=False)
```

```python
# ORDENAR POR EL SALARIO Y POR LA CIUDAD 
df.sort_values(by=['ciudad','salario'])
```

```python
df.sort_values(by=['ciudad','salario'], ascending=[True,True])
```

```python
df.sort_values(by=['ciudad','salario'], ascending=[True,False])
```

```python
df.sort_values(by=['ciudad','salario'], ascending=[False,False])
```
