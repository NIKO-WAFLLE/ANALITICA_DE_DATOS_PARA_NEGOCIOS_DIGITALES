# filtrado por operadores logico 

Los operadores lógicos permiten filtrar datos combinando condiciones.

| **Operador** | **Significado** | **¿Qué hace?**                                               |
|--------------|-----------------|--------------------------------------------------------------|
| `&`          | AND (Y)         | Devuelve filas que cumplen **todas las condiciones**        |
| `\|`          | OR (O)          | Devuelve filas que cumplen **al menos una condición**       |
| `~`          | NOT (NO)        | Devuelve filas que **no cumplen** una condición             |


> **Nota Importante**: Pon cada condición entre paréntesis. Usa `&`, `|` y `~` — no uses `and`, `or`, `not`.

# Sintaxis de los Operadores Lógicos

| **Operador**   | **Uso en Pandas**                | **Significado**                                 |
|----------------|----------------------------------|-------------------------------------------------|
| **Y** lógico       | `(condición1) & (condición2)`   | Ambas condiciones deben cumplirse             |
|**O** lógico       | `(condición1) \| (condición2)`  | Al menos una condición se cumple             |
| **NO** lógico      | `~(condición)`                  | Niega una condición (no se cumple)           |

## Tabla de sintaxis de cambios de datos

| **Conversión**         | **Sintaxis recomendada**                                                                 |
|--------------------------|-----------------------------------------------------------------------------------------|
| `object ➝ datetime`     | `df_join['columna'] = pd.to_datetime(df_join['columna'])`                               |
| `object ➝ float`        | `pd.to_numeric(df['col'], errors='coerce').astype(float)`                               |
| `object ➝ int` (sin NaN) | `pd.to_numeric(df['col'], errors='coerce').astype(int)`                                 |
| `object ➝ int` (con NaN) | `pd.to_numeric(df['col'], errors='coerce').astype('Int64')`                             |
